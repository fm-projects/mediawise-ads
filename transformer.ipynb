# Transformer

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
```

```python
df_norm
```

```python
transformer_df = df_norm[['points', 'value']]
```

```python
transformer_df['new_points'] = transformer_df['points'].apply(lambda points: [[float(point['lat']), float(point['lon']), point['azimuth']] for point in points])
```

```python
transformer_df
```

```python
transformer_df['new_points'][0]
```

```python
torch.cat([torch.tensor(transformer_df['new_points'][0]), torch.zeros((10, 3))])
```

```python
class PointDataset(Dataset):
    def __init__(self):
        self.points = [torch.cat([torch.tensor(i), torch.zeros((1024 - len(i), 3))]) for i in transformer_df['new_points']]
        self.targets = transformer_df['value'].values


    def __len__(self):
        return len(self.points)

    def __getitem__(self, idx):
        return self.points[idx], torch.tensor(self.targets[idx], dtype=torch.float32)
```

```python
dataset = PointDataset()
```

```python
dataloader = DataLoader(dataset, batch_size=1)
```

```python
next(iter(dataloader))
```

```python
encoder_layer = nn.TransformerEncoderLayer(d_model=3, nhead=3)
transformer_encoder = nn.TransformerEncoder(encoder_layer, num_layers=6)
src = torch.rand(20, 16, 3)
out = transformer_encoder(src)
out.shape
```

```python
class TransformerMLP(nn.Module):
    def __init__(self):
        super(TransformerMLP, self).__init__()
        self.encoder_layer = nn.TransformerEncoderLayer(d_model=3, nhead=3)
        self.transformer_encoder = nn.TransformerEncoder(encoder_layer, num_layers=1)

        self.encoder_MLP = nn.Sequential(nn.Linear(1024 * 3, 512),
                                         nn.ReLU(),
                                         nn.Linear(512, 128),
                                         nn.Sigmoid())

        self.decoder_MLP = nn.Sequential(nn.Linear(128, 1))

    def forward(self, x):
        self.emb = self.encoder_MLP(torch.flatten(self.transformer_encoder(x)))
        return self.decoder_MLP(self.emb)
```

```python
transformer_mlp = TransformerMLP()
```

```python
src = torch.rand(1, 1024, 3)
out = transformer_mlp(src)
out.shape
```

```python
criterion = nn.MSELoss()
optimizer = optim.Adam(transformer_mlp.parameters(), lr=4e-3, betas=(0.9, 0.98), eps=1e-9)

transformer_mlp.train()
for epoch in range(5):
    loss_sum = 0
    for i, batch in enumerate(dataloader):
        data, labels = batch
        data, labels = data, labels
        optimizer.zero_grad()

        output = transformer_mlp(data)
        loss = criterion(output, labels)
        loss.backward()
        loss_sum += loss.item()
        optimizer.step()
    print(epoch, loss_sum/len(dataloader))
transformer_mlp.eval();
```
